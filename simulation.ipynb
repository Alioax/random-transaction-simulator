```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

import numpy as np
import random

accounts = np.ones(50)*100

transactions_count = 10**6
transactions_exchange = 20


def transaction(accounts=accounts, transaction_exchange=transactions_exchange):
    giver = random.randrange(50)
    receiver = random.randrange(50)

    while receiver == giver:
        receiver = random.randrange(50)

    while accounts[giver] < transaction_exchange:
        giver = random.randrange(50)

    accounts[giver] -= transaction_exchange
    accounts[receiver] += transaction_exchange


for i in range(0, transactions_count):
    if (i % (transactions_count/10) == 0):

        colors = np.random.rand(50)

        plt.scatter(range(0, 50), np.sort(accounts), c=colors)
        plt.ylabel(
            f'wealth after {i} transaction{"s" if i > 0 else ""} of {transactions_exchange}$')
        plt.xlabel(f'account ID')
        plt.show()
        # plt.savefig(f'{i} transactions', dpi=512, format=None, metadata=None,
        # bbox_inches=None, pad_inches=0.1,
        # facecolor='auto', edgecolor='auto',
        # backend=None,
    #    )
        # plt.clf()
    transaction()
```
